```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('Ecommerce Customers')
```

```python
df.info()
```

```python
sns.jointplot(x='Time on Website',y='Yearly Amount Spent',data=df)
```

```python
sns.jointplot(x='Time on App',y='Length of Membership',data=df,kind='hex')
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#instantiated the estimator
model=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
sns.lmplot(x='Yearly Amount Spent',y='Length of Membership',data=df)
```

```python
df.columns
```

```python
X=df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
y=df['Yearly Amount Spent']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
plt.scatter(predictions,y_test)
```

```python
plt.hist(predictions-y_test)
```

```python
sns.distplot(predictions-y_test,bins=50)
```

```python
sns.kdeplot(predictions-y_test)
```

```python
#evaluating performance
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(predictions,y_test)
```

```python
metrics.mean_squared_error(predictions,y_test)
```

```python
np.sqrt(metrics.mean_squared_error(predictions,y_test))
```
