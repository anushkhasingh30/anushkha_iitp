```python
from random import shuffle
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
```

```python
class card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
    def __str__(self):
        print(f'{self.rank} of {self.suit}')
```

```python
class deck:
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(card(suit,rank))
     
    def __str__(self):
        for item in self.all_cards:
            item.__str__()
  
    def shuffle(self):
        shuffle(self.all_cards)
        
    def deal(self):
        return self.all_cards.pop()
            
```

```python
class Hand:
    def __init__(self):
        self.cards=[]
        self.value=0
        self.aces=0
        
    def add_cards(self,card):
        self.cards.append(card)
        self.value+=card.value
        if card.rank=='Ace':
            self.aces+=1
            
    def adjust_for_aces(self):
        while self.value>21 and self.aces:
            self.value-=10
            self.aces-=1
            
    def __str__(self):
        self.cards[-1].__str__()
```

```python
class chips:
    def __init__(self):
        self.total=100
        self.bet=0
        
    def win_bet(self):
        self.total+=self.bet
        
    def lose_bet(self):
        self.total-=self.bet
```

```python
def take_bet(player):
    while True:
        try:
            #this player is the chips object
            player.bet=int(input('Enter the players bet: '))
            
        except:
            print('pls provide an integer')
            
        else:
            if player.bet>player.total:
                print(f'sry! you dont have enough chips!! you only have {player.total} chips')
            else:
                break
```

```python
def hit(deck,Hand):
    Hand.add_cards(deck.deal())
    Hand.adjust_for_aces()
```

```python
def hit_or_stand(deck,Hand):
    playing=True
    while playing:
        x=input('hit or stand? h or s')
        if x[0]=='h':
            hit(deck,Hand)
        elif x[0]=='s':
            print('now its dealers turn')
            playing=False
        #else:
        #    print('i dont understand')
        break
```
